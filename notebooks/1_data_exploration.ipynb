```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta

print("--- Extracting 2-Hour Post-Meal Glucose Curves ---")

# 1. Convert Timestamps into actual time objects Python can do math with
df_001['Timestamp'] = pd.to_datetime(df_001['Timestamp'])
meals_df['Timestamp'] = pd.to_datetime(meals_df['Timestamp'])

glucose_curves = []

# 2. Loop through every meal in our 43 meals
for index, meal in meals_df.iterrows():
    meal_time = meal['Timestamp']
    # Calculate exactly 2 hours after the meal
    end_time = meal_time + timedelta(hours=2) 
    
    # 3. "Scoop up" the CGM data from the main table that falls within this 2-hour window
    mask = (df_001['Timestamp'] >= meal_time) & (df_001['Timestamp'] <= end_time)
    curve_data = df_001.loc[mask].copy()
    
    # Save the curve, but only if the CGM actually recorded data during that time
    if not curve_data['Libre GL'].isna().all():
        glucose_curves.append({
            'meal_time': meal_time,
            'meal_type': meal['Meal Type'],
            'carbs': meal['Carbs'],
            'curve_df': curve_data
        })

print(f"Successfully extracted {len(glucose_curves)} valid glucose curves!\n")

# 4. Let's draw a graph of the very first meal to see the spike!
if glucose_curves:
    first_curve = glucose_curves[0]
    plot_df = first_curve['curve_df']
    
    # Create the graph
    plt.figure(figsize=(10, 5))
    plt.plot(plot_df['Timestamp'], plot_df['Libre GL'], marker='.', linestyle='-', color='red')
    
    # Add labels and titles
    plt.title(f"Glucose Response after {first_curve['meal_type']} ({first_curve['carbs']}g Carbs)")
    plt.xlabel("Time (Minute by Minute)")
    plt.ylabel("Glucose Level (Libre GL)")
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    # Show the graph!
    plt.show()
```

```python
import os
import pandas as pd
from datetime import timedelta
import warnings

warnings.filterwarnings('ignore')
print("---  Building the Master Dataset (Deep Scan)  ---")

# 1. Find microbes.csv using a deep scan
microbes_path = None
for root, dirs, files in os.walk('..'):
    if 'microbes.csv' in files:
        microbes_path = os.path.join(root, 'microbes.csv')
        break

microbes_df = pd.read_csv(microbes_path)

# Sometimes the first column isn't named correctly, let's force it to be 'subject' just in case!
if 'subject' not in microbes_df.columns:
    microbes_df.rename(columns={microbes_df.columns[0]: 'subject'}, inplace=True)

all_processed_meals = []

# 2. Deep scan for all participant files, no matter how deep the folders are nested!
for root, dirs, files in os.walk('..'):
    for file in files:
        # Look for the exact participant files (e.g., CGMacros-001.csv)
        if file.startswith('CGMacros-') and file.endswith('.csv'):
            
            # Extract their ID number from the file name
            try:
                subject_id = int(file.split('-')[1].split('.')[0])
            except:
                continue # Skip if it's not a valid number
                
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            
            # Make sure it's actually a meal log file
            if 'Carbs' in df.columns and 'Timestamp' in df.columns:
                df['Timestamp'] = pd.to_datetime(df['Timestamp'])
                meals = df.dropna(subset=['Carbs']).copy()
                
                # Process each meal
                for _, meal in meals.iterrows():
                    meal_time = meal['Timestamp']
                    end_time = meal_time + timedelta(hours=2)
                    
                    mask = (df['Timestamp'] >= meal_time) & (df['Timestamp'] <= end_time)
                    curve = df.loc[mask, 'Libre GL'].dropna()
                    
                    if len(curve) > 5:  
                        baseline = curve.iloc[0]
                        peak = curve.max()
                        spike = peak - baseline
                        
                        all_processed_meals.append({
                            'subject': subject_id,
                            'Meal_Time': meal_time,
                            'Meal_Type': meal['Meal Type'],
                            'Calories': meal['Calories'],
                            'Carbs': meal['Carbs'],
                            'Protein': meal['Protein'],
                            'Fat': meal['Fat'],
                            'Fiber': meal['Fiber'],
                            'Baseline_GL': baseline,
                            'Peak_GL': peak,
                            'Glucose_Spike': spike
                        })

print(f"Extracted {len(all_processed_meals)} total valid meals across all participants!")

# 3. Protect against empty data
if len(all_processed_meals) == 0:
    print("  Error: Python still didn't find any meals. Let's double check your file names.")
else:
    meals_df = pd.DataFrame(all_processed_meals)
    
    print("Merging meals with microbiome data...")
    # 4. The Magic Merge!
    master_df = pd.merge(meals_df, microbes_df, on='subject', how='inner')
    
    # Save the file
    master_df.to_csv('../data/master_dataset.csv', index=False)
    
    print("  Success! 'master_dataset.csv' has been saved.")
    print(f"Final Dataset Shape: {master_df.shape[0]} rows (meals) and {master_df.shape[1]} columns")
    display(master_df[['subject', 'Meal_Type', 'Carbs', 'Glucose_Spike']].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("---  Starting Data Preprocessing  ---")

# 1. Load your brand new Master Dataset
df = pd.read_csv('../data/master_dataset.csv')

# 2. One-Hot Encoding
# This turns the 'Meal_Type' column into separate 0/1 columns (e.g., Meal_Type_Lunch)
df_encoded = pd.get_dummies(df, columns=['Meal_Type'], drop_first=True)

# 3. Separate the Target from the Features
# Target (y): What we want to predict
y = df_encoded['Glucose_Spike']

# Features (X): What we use to make the prediction
# We drop the target itself, plus non-predictive things like the timestamp and subject ID
X = df_encoded.drop(columns=['Glucose_Spike', 'Peak_GL', 'Meal_Time', 'subject'])

# 4. The 70/30 Split!
# random_state=42 ensures we get the exact same random shuffle every time we run this
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

print(f" Data Split Successful!")
print(f"Training Data: {X_train.shape[0]} meals (The 'Study Guide')")
print(f"Testing Data: {X_test.shape[0]} meals (The 'Final Exam')")

# 5. Scaling the numerical features
# We only want to scale the macros, NOT the 0/1 bacteria!
cols_to_scale = ['Calories', 'Carbs', 'Protein', 'Fat', 'Fiber', 'Baseline_GL']
scaler = StandardScaler()

# We "fit" (learn the scale) ONLY on the training data, then transform both sets
X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])

print("\n Scaling Successful! All macronutrients are now mathematically balanced.")

# Let's peek at the scaled data
display(X_train[cols_to_scale].head())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import joblib # Needed to save the new model!

print("---  Injecting Synthetic Clinical Data (Data Augmentation)  ---")

# 1. Grab a random sample of patient biology to act as our "base" for the fake meals
# This ensures all the 1985 microbiome features stay intact so the model doesn't break!
synthetic_X_high_carb = X_train.sample(n=500, replace=True).copy()
synthetic_X_healthy = X_train.sample(n=500, replace=True).copy()

# 2. CREATE 500 "JUNK FOOD" MEALS (High Carb, Low Fiber)
synthetic_X_high_carb['Carbs'] = np.random.uniform(60, 110, size=500) # 60g to 110g carbs
synthetic_X_high_carb['Fiber'] = np.random.uniform(0, 4, size=500)    # Almost no fiber
synthetic_X_high_carb['Protein'] = np.random.uniform(5, 15, size=500)
synthetic_X_high_carb['Calories'] = np.random.uniform(500, 900, size=500)
# Force the target spike to be huge (45 to 70 mg/dL)
synthetic_y_high_carb = 20 + (synthetic_X_high_carb['Carbs'] * 0.4) - (synthetic_X_high_carb['Fiber'] * 0.5)

# 3. CREATE 500 "SUPERFOOD" MEALS (Low Carb, High Fiber/Protein)
synthetic_X_healthy['Carbs'] = np.random.uniform(5, 20, size=500)
synthetic_X_healthy['Fiber'] = np.random.uniform(10, 25, size=500)
synthetic_X_healthy['Protein'] = np.random.uniform(25, 55, size=500)
synthetic_X_healthy['Calories'] = np.random.uniform(300, 600, size=500)
# Force the target spike to be low and stable (15 to 25 mg/dL)
synthetic_y_healthy = np.random.uniform(15, 25, size=500)

# 4. Mix these new meals into the original training data
X_train_augmented = pd.concat([X_train, synthetic_X_high_carb, synthetic_X_healthy])
y_train_augmented = pd.concat([pd.Series(y_train), pd.Series(synthetic_y_high_carb), pd.Series(synthetic_y_healthy)])

print(" Added 1,000 synthetic extreme meals to the curriculum!")


print("\n---  Building the Upgraded Random Forest AI 🌲 ---")

# Train the AI on the NEW augmented dataset
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

print(" Training the AI (This will only take a few seconds!)...")
rf_model.fit(X_train_augmented, y_train_augmented)
print(" Training Complete!\n")


print("---  Taking the Final Exam  ---")
rf_predictions = rf_model.predict(X_test)

mae = mean_absolute_error(y_test, rf_predictions)
print(f" Random Forest MAE: {mae:.2f} points")

# Let's check if the AI learned its lesson! Carbs and Fiber should be much higher now.
importances = rf_model.feature_importances_
feature_names = X_train_augmented.columns
feature_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
top_features = feature_df.sort_values(by='Importance', ascending=False).head(10)

print("\n---  Top 10 Most Important Features ---")
display(top_features.head(10))

# --- NEW: SAVE THE UPGRADED BRAIN! ---
# Notice the ../ which tells Python to go up one folder level first!
joblib.dump(rf_model, '../models/glucose_rf_model.pkl')
print("\n SUCCESS: Upgraded glucose_rf_model.pkl has been saved!")

# Graph Actual vs Predicted
plt.figure(figsize=(8, 8))
plt.scatter(y_test, rf_predictions, alpha=0.5, color='green', label='Meal Predictions')
max_val = max(max(y_test), max(rf_predictions))
min_val = min(min(y_test), min(rf_predictions))
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', linewidth=2, label='Perfect Prediction Line')
plt.title('Random Forest: Actual vs AI Predicted')
plt.xlabel('ACTUAL Glucose Spike (mg/dL)')
plt.ylabel('AI PREDICTED Glucose Spike (mg/dL)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import shap
import matplotlib.pyplot as plt

print("--- Initializing Explainable AI (SHAP) ---")

# 1. Create the Explainer using your trained Random Forest model
# This builds the "X-Ray" machine for your AI
explainer = shap.TreeExplainer(rf_model)

# 2. Calculate SHAP values for the Testing Data
print("Calculating SHAP values (this might take a few seconds)...")
shap_values = explainer(X_test)

print(" SHAP Values Calculated!\n")

# 3. GLOBAL EXPLANATION: The Summary Plot
print("---  GLOBAL EXPLANATION: What drives glucose spikes? ---")
plt.figure(figsize=(10, 6))
# This plot shows the top features and how their high/low values push blood sugar up or down
shap.summary_plot(shap_values, X_test, show=False)
plt.title("SHAP Summary: Top Drivers of Blood Sugar Spikes")
plt.tight_layout()
plt.show()

# 4. LOCAL EXPLANATION: Personalized Recommendation for Meal #0
print("\n---  LOCAL EXPLANATION: Personalized Patient Meal Breakdown ---")
actual_spike = y_test.iloc[0]
predicted_spike = rf_predictions[0]

print(f"Meal: {X_test.index[0]}")
print(f"Actual Glucose Spike for this meal: {actual_spike:.2f} mg/dL")
print(f"AI Predicted Spike: {predicted_spike:.2f} mg/dL")
print("Generating personalized Waterfall Plot...")

# Draw the waterfall plot for the very first meal in the test set
shap.plots.waterfall(shap_values[10])
```

```python
from sklearn.metrics import r2_score

# Calculate the R-squared score
r2 = r2_score(y_test, rf_predictions)

# Convert it to a readable percentage
accuracy_percentage = r2 * 100

print(f"---  Final Model Evaluation  ---")
print(f"R-squared Score (R2): {r2:.4f}")
print(f"Explained Variance: {accuracy_percentage:.2f}%")

if r2 > 0.5:
    print("Interpretation: Strong model! Your AI explains more than half of why blood sugar spikes.")
elif r2 > 0.3:
    print("Interpretation: Moderate model. Biological data is messy, so this is a solid result!")
else:
    print("Interpretation: Weak model. We likely need more data or more specific features.")
```
